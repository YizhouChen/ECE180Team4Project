```python
import json
import math
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
business = pd.read_json('business.json', lines=True)
g = sns.FacetGrid(data=business, col='stars')
g.map(plt.hist, 'review_count', bins=5)
plt.show()
```

```python
sns.boxplot(x='stars', y='review_count', data=business)
plt.show()
```

```python
sns.distplot(business.stars, kde=False)
plt.show()
```
